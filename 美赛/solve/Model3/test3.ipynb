```python
import pandas as pd
data = pd.read_excel('第四问.xlsx')

#data#.plot.line()
ret = pd.DataFrame(index=data.iloc[:,0].values, data = data.iloc[:,1:].values, columns=['Female susceptibility', 'Female infection', 'Male susceptibility', 'Male infection', 'Prey'])
```

```python
data
```

```python
import pylab as plt
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(ret.index, ret['Female susceptibility'], 'green', linestyle = '--', label='Female susceptibility')
plt.plot(ret.index, ret['Female infection'], 'green', label='Female infection')
plt.plot(ret.index, ret['Male susceptibility'], 'blue', linestyle = '--', label='Male susceptibility')
plt.plot(ret.index, ret['Male infection'], 'blue', label='Male infection')
plt.plot(ret.index, ret['Prey'], 'red', label='Prey')
plt.xlabel('Time')
plt.ylabel('Number')
plt.legend()
plt.subplot(2,1,2)
plt.plot(ret.index, (ret['Male infection'] + ret['Male susceptibility']) / (ret['Female infection'] + ret['Female susceptibility'] + ret['Male infection'] + ret['Male susceptibility']))
plt.xlabel('Time')
plt.ylabel('Male proportion')
plt.savefig(fname = 'four.jpg', dpi = 500, bbox_inches = 'tight')
```
