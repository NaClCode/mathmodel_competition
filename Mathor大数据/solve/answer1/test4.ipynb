```python
import pandas as pd
import lightgbm
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
data = pd.read_csv('time/time0.csv')
model = lightgbm.LGBMRegressor()
x = np.arange(0, 166).reshape(-1, 1)
y = data.qty
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2)
model.fit(train_x, train_y)
ret = {'trainR':r2_score(train_y, model.predict(train_x)), 'testR':r2_score(test_y, model.predict(test_x))}
pickle.dump(model, open('time0.pkl', 'wb'))
```
