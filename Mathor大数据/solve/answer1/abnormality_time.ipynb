```python
import pandas as pd
data = pd.read_csv('time/time16.csv')
```

```python
import seaborn as sns
import pylab as plt
from matplotlib.ticker import MaxNLocator
sns.set(style="darkgrid")
sns.lineplot(x="date", y="qty", data=data)
plt.plot(7, data.iloc[7, :].qty, 'ro')
plt.plot(8, data.iloc[8, :].qty, 'ro')
plt.plot(9, data.iloc[9, :].qty, 'ro')
plt.plot(12, data.iloc[12, :].qty, 'ro')
ax = plt.gca()
sns.despine()
ax.xaxis.set_major_locator(MaxNLocator(steps=[2]))
plt.xticks(rotation=40)

plt.savefig(fname = '异常值.png', dpi = 500, bbox_inches = 'tight', pad_inches = 0.0)
```
