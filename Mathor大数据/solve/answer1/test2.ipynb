```python
import pandas as pd
from sktime.forecasting.fbprophet import Prophet
data = pd.read_csv('time/time0.csv', index_col=4).sort_index().qty
model = Prophet(seasonality_mode='multiplicative')
model.fit(data.values)
```

```python
import numpy as np
yy = model.predict(np.arange(0, 166))
import pylab as plt
data.plot.line()
plt.plot(np.arange(0, 166), yy, 'r')
```

```python
from sktime.forecasting.trend import TrendForecaster
model = TrendForecaster()
model.fit(data.values)
```

```python
import numpy as np
yy = model.predict(np.arange(0, 166))
import pylab as plt
data.plot.line()
plt.plot(np.arange(0, 166), yy, 'r')
```

```python
import lightgbm
model = lightgbm.LGBMRegressor()
model.fit(np.arange(0, 166).reshape(-1, 1), data.values)
import numpy as np
yy = model.predict(np.arange(0, 166).reshape(-1, 1))
import pylab as plt
data.plot.line()
plt.plot(np.arange(0, 166), yy, 'r')
```
