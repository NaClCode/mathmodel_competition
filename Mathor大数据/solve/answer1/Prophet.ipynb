```python
import prophet
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
data = pd.read_csv('No_abnormality_time/time2.csv', index_col=3)['qty']
data = pd.DataFrame({'ds': pd.to_datetime(data.index), 'y': data.values})
length = data.shape[0]
train_data = data[:int(length * 0.8)]
test_data = data[int(length * 0.8):]
```

```python
md = prophet.Prophet()
md.add_country_holidays(country_name='CN')
md.fit(train_data)
```

```python
import pylab as plt
import numpy as np
def wmapes(y_true, y_pred):
    return 1 - np.sum(np.abs(y_true - y_pred)) / np.sum(y_true)
```

```python
test_data_predict = md.predict(df = test_data)#['yhat']
```

```python
wmapes(test_data.y.values, test_data_predict)
```
