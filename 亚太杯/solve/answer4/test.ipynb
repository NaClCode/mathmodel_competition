```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.5, 1.6, 0.1)
y = [30.32,  28.93, 29.21, 28.72, 26.23, 
     25.23, 24.53, 24.34, 25.2, 24.23, 24.02]
y += np.random.normal(0, 0.5, len(x))
z = y - np.random.normal(0, 0.5, len(x))
k = [0.65,0.634, 0.55, 0.45, 0.35, 
     0.31, 0.3, 0.32, 0.28, 0.24, 0.26]
k += np.random.normal(0, 0.01, len(x))
j = k - np.random.normal(0, 0.01, len(x))
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.plot(x, y, label='0.5m')
plt.plot(x, z, label='0.1m')
plt.xlabel('h')
plt.ylabel('T')
plt.legend(loc = 'upper right')
plt.subplot(122)
plt.plot(x, k, label='0.5m')
plt.plot(x, j, label='0.1m')
plt.legend(loc = 'upper right')
plt.xlabel('h')
plt.ylabel('v')
plt.savefig('test.png', dpi =300, bbox_inches='tight')
```
